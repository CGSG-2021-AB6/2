```python
import matplotlib.pyplot as plt
from math import sqrt 

x = []
num = 100
fig, ax = plt.subplots()
for i in range(num):
    x.append(i / num )
y1 = [d**2 for d in x] 

x = []
for i in range(num):
    x.append(i / num )


    
y2 = [sqrt(d) for d in x]



ax.plot(x, y2, label='sqrt(x)')
ax.plot(x, y1, label='x^2')
ax.set_title('функции')
ax.set_xlabel('x') 
ax.set_ylabel('y')
ax.legend()
```
